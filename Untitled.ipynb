```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (10.0, 6.0) # set default size of plots
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'black'
```

```python
#plt.figure(linewidth=1)
plt.title('MPI Strong Run Time n = 500',fontsize = 18)
plt.ylabel('Time',fontsize =20)
plt.xlabel('Nodes/Processes',fontsize = 20)

x = [1,2,4,6,12,16,24,32]
y_mpi = [0.421927,0.352966,0.292309,0.234349,0.257548,0.504896,0.524612,0.416573]
y_openmp = [0.246217,0.213953,0.209844,0.238323,0.170922,0.172989,0.179248,0.133327]

plt.loglog(x,y_mpi, 'go--',basex = 2,basey=2 ,markersize = 20, linewidth=4,label='MPI runtime')
plt.loglog(x,y_openmp, 'k^--',basex = 2,basey=2 ,markersize = 20, linewidth=4,label='OpenMP runtime')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.ylim([2**(-3),2**(0)])
plt.legend(loc='upper right', fontsize= 20)
plt.savefig("mpi_strong_runtime.png")
plt.show()
```
